```python
df.dropna(inplace=True)
df
```

```python
df=df.iloc[[0,3,6,8,11,14,15,18,21,23,24,27,29,32,36,39,41,42,43,44,53],:]
dr={'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)':721080.64}
df=df.append(dr,ignore_index=True)
df
```

```python
b=sns.barplot(x='Fund allotted(in ₹crores)',y='Department /Ministry',data=df)
b.axes.set_title("Ministry Vs Fund_Allocation",fontsize=20)
b.set_xlabel("Fund allotted(in ₹crores)",fontsize=10)
b.set_ylabel("Department /Ministry",fontsize=10)
b.tick_params(labelsize=8)
plt.show()
```

```python
df.plot.kde()
plt.ylabel('Max Fund allotted(in ₹crores)')
plt.show()
```

```python
data= df["Fund allotted(in ₹crores)"]
labels = df["Department /Ministry"]
plt.figure(figsize=(28,28))
plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=18)
plt.title("Distribution of Budget", fontsize=30)
plt.show()
```

The Analysis depicts that Ministry of Finance has highest allocation with 35.3%.
Followed by Ministry of Defence with allocation of 12.2% & Ministry of Consumers Affairs 6.5%.
